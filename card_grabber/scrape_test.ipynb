```python
# import dependencies

import requests
from PIL import Image
from io import BytesIO
import pandas as pd
```

```python
# Save image location as a variable
img_url = "https://limitlesstcg.nyc3.cdn.digitaloceanspaces.com/pocket/A1/A1_001_EN_SM.webp"

# Download the image from the URL (Python doesn't know this is an image yet. It's just a blob of data)
response = requests.get(img_url)
```

```python
# Open and identiify an image file and return an Image Object
image = Image.open(
    # BytesIO wraps the raw bytes ("blob of data" from request) into something that behaves like a file on disk
    BytesIO(response.content)
    )
```

```python
# render image to verify that it worked properly
image
```

```python
# Close image so it doesn't waste energy and resources
image.close()
# Can't run pytesseract on closed image
```
